```python
from sklearn.svm import SVC
from sklearn.preprocessing import Imputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, classification_report
from sklearn.model_selection import train_test_split

imp = Imputer(missing_values = 'NaN', strategy = 'most_frequent', axis=0)
clf = SVC(decision_function_shape='ovo', probability=True)

steps = [('imputation', imp), ('SVC', clf)]

pipeline = Pipeline(steps)

merged["Runtime (Minutes)"] = merged["Runtime (Minutes)"].astype(int)
merged["Votes"] = merged["Votes"].astype(int)
merged["Revenue (Millions)"] = merged["Revenue (Millions)"].astype(int)
merged["Metascore"] = merged["Metascore"].astype(int)
merged["Rating"] = merged["Rating"].astype(int)

xdata = merged[["Runtime (Minutes)", "Votes", "Revenue (Millions)", "Metascore"]].values
ydata = merged["Rating"].values

xtra, xtes, ytra, ytes = train_test_split(xdata, ydata, random_state=40, test_size=0.3)

pipeline.fit(xtra, ytra)
```

```python
ypred = pipeline.predict(xtes)

## R2 Score
print(pipeline.score(xtes, ytes))

## Mean-Squared Error
print(mean_squared_error(ytes, ypred))

## Classification Report
print(classification_report(ytes, ypred))
```

```python
from sklearn import linear_model

reg = linear_model.LassoLars(alpha=3)
reg.fit(xtra, ytra)

ypredLasso = reg.predict(xtes)
ypredLasso = ypredLasso.reshape(-1, 1).astype(int)

##print(ytes.shape)

## R2 Score
print(reg.score(xtes, ytes))

## Mean-Squared Error
print(mean_squared_error(ytes, ypredLasso))
```
